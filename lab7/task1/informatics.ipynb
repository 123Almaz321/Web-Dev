Циклы: Задачи на цикл while


```python
#A
import math
n = int(input())
i = 1
while i < n:
    if math.sqrt(i) % 1  == 0:
        print(i)
    i += 1
```

```python
#B
x = int(input())
i = 2
while i < x:
    if x % i == 0:
        print(i)
        break
    i += 1
```

```python
#C
n = int(input())
i = 0
while 2**i < n:
    print(2**i)
    i += 1
```

```python
#D
number = int(input())

if number <= 0:
    print("NO")
else:
    while number % 2 == 0:
        number //= 2
    
    if number == 1:
        print("YES")
    else:
        print("NO")
```

```python
#E
t = int(input())
c = 1
e = 0

while c < t:
    c *= 2
    e += 1

print(e)
```

Массивы: Условия задач


```python
#A
n = int(input())
arr = []
for i in range(n):
    a = int(input())
    arr.append(a)

for i in range(n):
    if i % 2 == 0:
        print(arr[i])
```

```python
#B
n = int(input())
arr = []
for i in range(n):
    a = int(input())
    arr.append(a)

for i in range(n):
    if arr[i] % 2 == 0:
        print(arr[i])
```

```python
#C
n = int(input())
arr = []
summ = 0
for i in range(n):
    a = int(input())
    arr.append(a)

for i in range(n):
    if arr[i] > 0:
        summ += 1
print(summ)
```

Функции: Функции


```python
#A
a = int(input())
b = int(input())
c = int(input())
d = int(input())

def minn(a, b, c, d):
    m = a
    if b < m:
        m = b
    if c < m:
        m = c
    if d < m:
        m = d
    return m
    

print(minn(a,b,c,d))
```

```python
#B
def degree(a, n):
    return a**n

a = int(input())
n = int(input())

print(degree(a,n))
```

```python
#C
def t(a, b):
    if not a and b or a and not b:
        return True
    else:
        return False
    
a = bool(input())
b = bool(input())
print(t(a, b))
```
